2.1 전체 기사의 형태소 분석

```python
import os
import re

from konlpy.tag import Okt

okt = Okt()

dirs = ['it-life-hack', 'movie-enter']

docterm = []

label = []

tmp1 = []
tmp2 = []

for i, d in enumerate(dirs):
    files = os.listdir('../data/'+ d)
    #print(file)

    for file in files:
        f = open('../data/' + d +'/' + file, 'r', encoding='utf-8')
        text = f.read()
        #print(text)
        req_text = re.sub(r'[0-9a-zA-Z]+', '', text)
        req_text = re.sub(r'[:;/+\.-]', '', req_text)
        req_text = re.sub(r'[\s\n]', '', req_text)

        for token in okt.nouns(req_text):
            tmp1.append(token)
            tmp2 = ' '.join(tmp1)
        docterm.append(tmp2)

        tmp1 = []

        label.append(i)
        f.close()
```

2.2 단어 문서 행렬의 작성

```python
#일반적으로 어떤 문서 중에 포함된 각 단어의 출현 빈도를 표 형식으로 만든다. 그것이 단어 문서 행렬이라고 지칭한다.
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np

cv = CountVectorizer()

docterm_cv = cv.fit_transform(np.array(docterm))
docterm_cnt = docterm_cv.toarray()
#print(docterm_cnt)

#docterm_cv


text_df_word =pd.DataFrame(docterm_cnt)

text_df_word.head()
```

```python
print(text_df_word.shape)


print(cv.get_feature_names()[0:50])
```

```python
# 연습문제 3 책에서 만든 풀이
wcnt = []

# 단어 출현회수를 카운트
docterm_wcnt = np.sum(a=docterm_cnt, axis=0)
# 단어와 그 출현회수를 하나씩 취득
for w, cnt in zip(cv.get_feature_names(), docterm_wcnt):
    wcnt.append([w, cnt])

# 데이터 프레임으로 변환하고 내림차순으로 소트
wcnt_df = pd.DataFrame(wcnt)
wcnt_df = wcnt_df.sort_values(1, ascending=False)
wcnt_df.head()
```
